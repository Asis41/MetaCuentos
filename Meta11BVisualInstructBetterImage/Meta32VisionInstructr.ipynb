```python
import requests
import torch
from PIL import Image
from transformers import MllamaForConditionalGeneration, AutoProcessor
import os

model_id = "meta-llama/Llama-3.2-11B-Vision-Instruct"

model = MllamaForConditionalGeneration.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
processor = AutoProcessor.from_pretrained(model_id)

# Ruta de la carpeta con las imágenes
carpeta_imagenes = 'Images_1'

# Extensiones de archivos de imagen que deseas incluir
extensiones_imagenes = ['.png', '.jpg', '.jpeg', '.bmp', '.gif']

# Inicializa un contador
contador = 1

prompt_lst = []

# Recorre los archivos en la carpeta
for archivo in os.listdir(carpeta_imagenes):
    # Verifica si el archivo tiene una extensión de imagen
    if any(archivo.lower().endswith(ext) for ext in extensiones_imagenes):
        image = Image.open(f"Images_1/{archivo}")

        messages = [
            {"role": "user", "content": [
                {"type": "image"},
                {"type": "text", "text": "Describe la imagen"}
            ]}
        ]
        input_text = processor.apply_chat_template(messages, add_generation_prompt=True)
        inputs = processor(
            image,
            input_text,
            add_special_tokens=False,
            return_tensors="pt"
        ).to(model.device)
        
        output = model.generate(**inputs, max_new_tokens=500)        
        prompt_lst.append(f"[{contador}.-" + processor.decode(output[0])[144:(len(processor.decode(output[0])) - 10)])
        contador += 1
print("Terminado")
```
